```python
#defaul libraries
#https://docs.python.org/ja/
import math
import random

import numpy as np #https://numpy.org/
import pandas as pd #https://pandas.pydata.org/

import sklearn #https://scikit-learn.org/stable/

import matplotlib.pyplot as plt #https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html
import seaborn as sns
```

```python
# データ生成
x = np.linspace(0, math.pi*2, 1000)

# プロット領域(Figure, Axes)の初期化
plt.figure()
plt.plot(x,np.sin(x),color='green',linestyle='--')
plt.show()
```

```python
# データ生成
x = np.linspace(0, math.pi*2, 1000)
y=[0]*1000
for i in range(1000):y[i]=random.randint(0,i)

# プロット領域(Figure, Axes)の初期化
plt.figure()
plt.scatter(x,y,color='green',marker='o',s=5)
plt.show()
```

```python
# データ生成
from sklearn.datasets import load_iris

iris=load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df['species']=['setosa']*50+['versicolor']*50+['virginica']*50
```

```python
sns.pairplot(df, hue='species')
```

```python
sns.heatmap(abs(df.corr()), square=True, annot=True,cmap='bone')
```

```python
sns.lineplot(data=df)
```

```python
sns.barplot(data=df)
```

```python
fig=plt.figure(figsize=(15,5))
sns.set(font_scale=0.7) 

ax=fig.add_subplot(1,3,1)
sns.barplot(data=df[df['species']=='setosa'],color='red')
ax=fig.add_subplot(1,3,2)
sns.barplot(data=df[df['species']=='versicolor'],color='green')
ax=fig.add_subplot(1,3,3)
sns.barplot(data=df[df['species']=='virginica'],color='blue')
```
