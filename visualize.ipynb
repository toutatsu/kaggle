```python
#defaul libraries
#https://docs.python.org/ja/
import math
import random

import numpy as np #https://numpy.org/
import pandas as pd #https://pandas.pydata.org/

import sklearn #https://scikit-learn.org/stable/

import matplotlib.pyplot as plt #https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html
import seaborn as sns
```

```python
# データ生成
x = np.linspace(0, math.pi*2, 1000)

# プロット領域(Figure, Axes)の初期化
plt.figure()
plt.plot(x,np.sin(x),color='green',linestyle='--')
plt.show()
```

```python
# データ生成
x = np.linspace(0, math.pi*2, 1000)
y=[0]*1000
for i in range(1000):y[i]=random.randint(0,i)

# プロット領域(Figure, Axes)の初期化
plt.figure()
plt.scatter(x,y,color='green',marker='o',s=5)
plt.show()
```

```python
# データ生成
from sklearn.datasets import load_iris

iris=load_iris()
df=pd.DataFrame(iris.data, columns=iris.feature_names)
df['species']=['setosa']*50+['versicolor']*50+['virginica']*50
```

```python
sns.pairplot(df, hue='species')
```

```python
sns.heatmap(abs(df.corr()), square=True, annot=True,cmap='bone')
```

```python
sns.lineplot(data=df)
```

```python
sns.barplot(data=df)
```

```python
iris.feature_names
```

```python
fig=plt.figure(figsize=(15,5))
sns.set(font_scale=0.7) 

ax=fig.add_subplot(1,3,1)
sns.barplot(data=df[df['species']=='setosa'],color='red')
ax=fig.add_subplot(1,3,2)
sns.barplot(data=df[df['species']=='versicolor'],color='green')
ax=fig.add_subplot(1,3,3)
sns.barplot(data=df[df['species']=='virginica'],color='blue')
```

#　複数

```python
fig,axes=plt.subplots(
    figsize=(12,6),
    nrows=2,
    ncols=2,
    sharex=True,
    subplot_kw=dict(xlabel='$x$',ylabel='$y$')
)

x=np.arange(0,1,0.1)

axes[0,0].plot(x,np.sin(x),color='red',label='$y=\sin{x}$')
axes[0,1].plot(x,np.sin(x),color='red',label='$y=\sin{x}$')
axes[1,0].plot(x,np.sin(x),color='red',label='$y=\sin{x}$')
axes[1,1].plot(x,np.sin(x),color='red',label='$y=\sin{x}$')

#凡例
axes[0,0].legend(loc='best')
axes[0,1].legend(loc='best')
axes[1,0].legend(loc='best')
axes[1,1].legend(loc='best')

#レイアウトの自動調整

fig.tight_layout()
```

```python
# フォントを全て読み込み
import matplotlib
fonts = set([f.name for f in matplotlib.font_manager.fontManager.ttflist])
 
# 描画領域のサイズ調整
plt.figure(figsize=(10,len(fonts)/4))
 
# フォントの表示
for i, font in enumerate(fonts):
    plt.text(0, i, f"日本語：{font}", fontname=font)
    
# 見やすいように軸を消す
plt.ylim(0, len(fonts))
plt.axis("off")
    
plt.show()
```
